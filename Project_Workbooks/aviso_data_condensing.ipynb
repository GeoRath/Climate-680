#### Slim AVISO Data into Monthly Dataset

Will help ease the calculations for entirity of workbooks after exporting results into netcdf data

<div class="alert alert-block alert-info">
Import Necessary Modules
</div>

```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

<div class="alert alert-block alert-info">
Read in Full Dataset
</div>

```python
# Aviso Data
file_aviso = "/project/airsea/AVISO/1993_2015_15N-15S_AVISO_SLA_ADT.nc"
ds_aviso = xr.open_dataset(file_aviso)
ds_aviso
```

<div class="alert alert-block alert-info">
Resample into Monthly Data
</div>

```python
ds_aviso = ds_aviso['sla'].resample(time="1MS").mean(dim="time")
ds_aviso
```

<div class="alert alert-block alert-info">
Convert Monthly Data to .nc file
</div>

```python
ds_aviso.to_netcdf('monthly_aviso.nc')
```
