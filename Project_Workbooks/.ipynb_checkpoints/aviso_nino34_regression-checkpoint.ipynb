## Aviso and Nino 3.4 Linear Regression

<div class="alert alert-block alert-info">
Import Necessary Modules
</div>

```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np

import cartopy.crs as ccrs
import cartopy.mpl.ticker as cticker
from cartopy.util import add_cyclic_point
```

<div class="alert alert-block alert-info">
Import Data
</div>

```python
# Nino 3.4 Data
file_nino34='/scratch/kpegion/nino34_1982-2019.oisstv2_anoms.nc'
ds_nino34=xr.open_dataset(file_nino34)
ds_nino34
```

```python
# Aviso Data
ds_aviso = xr.open_dataset('monthly_aviso.nc')
ds_aviso
```

<div class="alert alert-block alert-info">
Prepare Data for Analysis
</div>

```python
# Slice so that times match
ds_nino34=ds_nino34.sel(time=slice(ds_aviso['time'][0],ds_aviso['time'][-1]))
ds_nino34
```

<div class="alert alert-block alert-info">
Retrieve Anomalies
</div>

```python
ds_climo=ds_aviso.groupby('time.month').mean()
ds_anoms=ds_aviso.groupby('time.month')-ds_climo
ds_anoms
```

### Linear Regression

```python
from scipy.stats import linregress
```

```python
# Define Lengths
nx = len(ds_anoms['longitude'])
ny = len(ds_anoms['latitude'])

# Create Array
p_array = np.zeros((ny,nx))
r_array = np.zeros((ny,nx))
m_array = np.zeros((ny,nx))

x = ds_nino34['sst']
```

```python
# Looping

for i in range(nx):
    for j in range(ny):
        
        y = ds_anoms['sla'][:,j,i]
        
        m,b,r,p,e = linregress(x,y)
        
        p_array[j,i]= p
        r_array[j,i] = r
        m_array[j,i] = m
```

<div class="alert alert-block alert-info">
Make a Mask
</div>

```python
mask_sig = np.where(p_array<0.05,m_array,np.nan)
```

```python
clevs=np.arange(-0.2,0.21,0.0125)

fig = plt.figure(figsize=(11,8.5))

# Set the axes using the specified map projection
ax=plt.axes(projection=ccrs.PlateCarree(central_longitude=180))

# Add cyclic point
data=m_array
data,lon=add_cyclic_point(mask_sig,coord=ds_anoms['longitude'])
data_mask,lon=add_cyclic_point(mask_sig,coord=ds_anoms['longitude'])

# Make a filled contour plot
cs=ax.contourf(lon,ds_anoms['latitude'],
            data,clevs,
            transform=ccrs.PlateCarree(),
            cmap='bwr',extend='both')

ax.contourf(lon,ds_anoms['latitude'],data_mask,clevs,
            transform=ccrs.PlateCarree(),
            cmap='bwr',extend='both')

# Add coastlines
ax.coastlines()

# Add gridlines
#ax.gridlines()

# Define the xticks for longtitude 
ax.set_xticks(np.arange(-180,181,60),crs=ccrs.PlateCarree())
lon_formatter=cticker.LongitudeFormatter()
ax.xaxis.set_major_formatter(lon_formatter)

# Define ytick for latitude
ax.set_yticks(np.arange(-15,16,5),crs=ccrs.PlateCarree())
lat_formatter=cticker.LatitudeFormatter()
ax.yaxis.set_major_formatter(lat_formatter)

# Call colorbar
cbar=plt.colorbar(cs,orientation='horizontal',shrink=0.7,
                 label='Regression Coefficient (m/degC)')

# Add title
plt.title('Regression between Nino3.4 and SLA Anomalies')
```
